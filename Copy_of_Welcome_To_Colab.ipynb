```
# Mapping example (you may need a separate file or dictionary)
country_map = {1: 'India', 14: 'Australia', 30: 'Brazil', 37: 'Canada', 94: 'Indonesia'}
df['Country'] = df['Country Code'].map(country_map)

# Distribution
country_counts = df['Country'].value_counts()
print(country_counts)
```

```
top_cities = df['City'].value_counts().head(10)
print(top_cities)
```

```
# Split and explode to count each cuisine separately
cuisine_series = df['Cuisines'].dropna().str.split(',').explode().str.strip()
top_cuisines = cuisine_series.value_counts().head(10)
print(top_cuisines)
```

```

import folium
from folium.plugins import MarkerCluster
import os
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=12)

marker_cluster = MarkerCluster().add_to(restaurant_map)
for _, row in df.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        tooltip=f"{row['Restaurant Name']} ({row['City']})"
    ).add_to(marker_cluster)
map_file_path = "/mnt/data/restaurant_map.html"

output_dir = os.path.dirname(map_file_path)
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

restaurant_map.save(map_file_path)

map_file_path
```

```
import matplotlib.pyplot as plt
import seaborn as sns


city_counts = df['City'].value_counts().head(15)

country_counts = df['Country Code'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=city_counts.values, y=city_counts.index, palette="viridis")
plt.title("Top 15 Cities by Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```
top_cities = df['City'].value_counts().head(15).index
city_avg_ratings = df[df['City'].isin(top_cities)].groupby('City')['Aggregate rating'].mean().sort_values()

plt.figure(figsize=(12, 6))
sns.barplot(x=city_avg_ratings.values, y=city_avg_ratings.index, palette='mako')
plt.title("Average Restaurant Rating by Top 15 Cities")
plt.xlabel("Average Rating")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

```
country_avg_ratings = df.groupby('Country Code')['Aggregate rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=country_avg_ratings.index.astype(str), y=country_avg_ratings.values, palette='crest')
plt.title("Average Restaurant Rating by Country Code")
plt.xlabel("Country Code")
plt.ylabel("Average Rating")
plt.tight_layout()
plt.show()
```

```
from scipy.stats import pearsonr

lat_corr, _ = pearsonr(df['Latitude'], df['Aggregate rating'])
lon_corr, _ = pearsonr(df['Longitude'], df['Aggregate rating'])

print(f"Correlation between Latitude and Rating: {lat_corr:.3f}")
print(f"Correlation between Longitude and Rating: {lon_corr:.3f}")
```
